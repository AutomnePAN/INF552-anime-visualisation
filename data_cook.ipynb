### Completion of `Starting season`

Firstly we consider dedicating the `Starting season` by `Start airing`

```python
def get_season_from_date(date_str):
    date = pd.to_datetime(date_str)
    month = date.month
    if month < 4:
        return 'Spring'
    if month < 7:
        return 'Summer'
    if month < 10:
        return 'Fall'
    else:
        return 'Winter'
```

```python
count = 0
for index, anime in data_anime.iterrows():
    if anime['Starting season'] == '-' and anime['Episodes'] != '1':
        data_anime.set_value(index, 'Starting season', 'Movie')
    elif anime['Starting season'] == '-' and anime['Start airing'] != '-':
        data_anime.set_value(index, 'Starting season', get_season_from_date(anime['Start airing']))
        count += 1
```

```python
for c in data_anime.columns:
    if '-' in data_anime[c].values:
        print("Column: ", c)
        print("percentage of lost data: %.3f"% ( ( data_anime[c].value_counts() / len(data_anime) )['-'] * 100), "%") # no need to drop
        print(64 * "-=")
        print()
```

We can see that the percentage of missing data of `Starting season` drop to $0.8\%$

### Completion of `End airing`

Knowing that japanese animations are mostly weekly updated, we could calculated the `End airing` by  `Start airing` $+$ 7 days * `Episodes` 

```python
def get_end_airing_from_start_airing(date_str, episodes):
    
    start_day = pd.to_datetime(date_str)
    end_day = start_day + pd.DateOffset( 7 * int(episodes) )
    return "%d-%d-%d"%(end_day.year, end_day.month, end_day.day)
```

```python
count = 0
for index, anime in data_anime.iterrows():
    if anime['Episodes'] != '-' and anime['Start airing'] != '-' and anime['End airing'] == '-':
        data_anime.set_value(index, 'End airing', get_end_airing_from_start_airing(anime['Start airing'], anime['Episodes']))
        count += 1
```

## Get time line of the anime broadcast

```python
start_airing_days = []
for index, anime in data_anime.iterrows():
    start_airing_days.append( pd.to_datetime(anime['Start airing']) )
    
```

```python
start_airing_days.sort()
```

```python
recorded_years = range( start_airing_days[0].year, start_airing_days[-1].year + 1 )
```

```python
genre_by_year = {}
```

```python
for y in recorded_years:
    genre_by_year[y] = { g : {"scores": [], "favorites": [], "scored_by": [], "animes": []} for g in genres }
```

```python
data_anime[]
```

```python
for index, anime in data_anime.iterrows():
    year =  pd.to_datetime(anime['Start airing']).year
    for genre in genres:
        if genre in anime['Genres'].split(","):
            genre_by_year[year][genre]['scores'].append(anime['Score'])
            genre_by_year[year][genre]['animes'].append(anime['Title'])
            genre_by_year[year][genre]['favorites'].append(anime['Favorites'])
            genre_by_year[year][genre]['scored_by'].append(anime['Scored by'])
```

### Get the average score by year

```python
for g in genres:
    for y in recorded_years:
        scored_by = np.array( genre_by_year[y][g]['scored_by'] )
        scores = np.array(genre_by_year[y][g]['scores'])
        average_score = np.average(scores, weights=scored_by) if len(scores) > 0 else 0
        average_favorite = sum(genre_by_year[y][g]['favorites'])/len(scores) if len(scores) > 0 else 0
        genre_by_year[y][g]['average_score'] = average_score
        genre_by_year[y][g]['average_favorite'] = average_favorite
        
```

```python
with open("./data/genre_by_year.json", "w") as write_file:
    json.dump(genre_by_year, write_file)
```

```python
genre_count = {}
for g in genres:
    genre_count[g] = 0 
    
    for y in recorded_years:
        genre_count[g] += len( genre_by_year[y][g]['animes'] )
```
