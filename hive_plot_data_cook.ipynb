```python
import pandas as pd
import json
import re
import collections
```

```python
data_anime_filtered = pd.read_csv("./data/data_anime_filtered.csv")
for index, anime in data_anime_filtered.iterrows():
    data_anime_filtered.at[index,'Genres'] = re.sub('\xa0Adventure', 'Adventure', anime['Genres'])
data_anime_filtered=data_anime_filtered.rename(columns = {'Unnamed: 0':'id'})
data_anime_filtered.sort_values(by=['Score'],ascending=False).head(100)
data_anime_filtered=data_anime_filtered[data_anime_filtered['Score'] >= 9]
data_anime_filtered
```

```python
data_anime_filtered.to_csv('./data/data_anime_hive.csv', index=False)
```

## Get genres json

```python
# genres count
genres_count = {}

for index, anime in data_anime_filtered.iterrows():
    for genre in anime['Genres'].split(','):
        if genre not in genres_count:
            genres_count[genre] = 1
        else:
            genres_count[genre] += 1

od = collections.OrderedDict(sorted(genres_count.items(), key=lambda item: item[1]))

genres_js = []
ind = 0
for key, value in od.items():
    genres_js.append({'id': ind, 'genre': key, 'count': value})
    ind += 1

with open("./data/genres.json", "w") as write_file:
    json.dump(genres_js, write_file)
```

```python
genre2ind = {}
for item in genres_js:
    genre2ind[item['genre']]=item['id']
genre2ind
```

## Get Studio json

```python
# studios count
studios_count = {}

for index, anime in data_anime_filtered.iterrows():
    if anime['Studios'] not in studios_count:
        studios_count[anime['Studios']] = 1
    else:
        studios_count[anime['Studios']] += 1

od = collections.OrderedDict(sorted(studios_count.items(), key=lambda item: item[1]))

studios_js = []
ind = 0
for key, value in od.items():
    studios_js.append({'id': ind, 'studio': key, 'count': value})
    ind += 1

with open("./data/studios.json", "w") as write_file:
    json.dump(studios_js, write_file)
```

```python
studio2ind = {}
for item in studios_js:
    studio2ind[item['studio']]=item['id']
studio2ind
```

## Get Source Json

```python
# source count
sources_count = {}

for index, anime in data_anime_filtered.iterrows():
    if anime['Sources'] not in sources_count:
        sources_count[anime['Sources']] = 1
    else:
        sources_count[anime['Sources']] += 1

od = collections.OrderedDict(sorted(sources_count.items(), key=lambda item: item[1]))

sources_js = []
ind = 0
for key, value in od.items():
    sources_js.append({'id': ind, 'source': key, 'count': value})
    ind += 1

with open("./data/sources.json", "w") as write_file:
    json.dump(sources_js, write_file)
```

```python
source2ind = {}
for item in sources_js:
    source2ind[item['source']]=item['id']
source2ind
```

```python
# rating count
ratings_count = {}

for index, anime in data_anime_filtered.iterrows():
    if anime['Rating'] not in ratings_count:
        ratings_count[anime['Rating']] = 1
    else:
        ratings_count[anime['Rating']] += 1
```

```python
anime_source = []
anime_studio = []
anime_genre = []
studio_genre = []

for index, anime in data_anime_filtered.iterrows():
    anime_source.append({'anime_id': index, 'source_id':source2ind[anime["Sources"]]})
    anime_studio.append({'anime_id': index, 'studio_id':studio2ind[anime["Studios"]]})
    for genre in anime['Genres'].split(','):
        anime_genre.append({'anime_id': index, 'genre_id':genre2ind[genre]})
        studio_genre.append({'studio_id':studio2ind[anime["Studios"]], 'genre_id':genre2ind[genre]})
```

```python
with open("./data/anime_source.json", "w") as write_file:
    json.dump(anime_source, write_file)
with open("./data/anime_studio.json", "w") as write_file:
    json.dump(anime_studio, write_file)
with open("./data/anime_genre.json", "w") as write_file:
    json.dump(anime_genre, write_file)
with open("./data/studio_genre.json", "w") as write_file:
    json.dump(studio_genre, write_file)
```
